## OOM (Object Oriented Method)

```python
fig, ax = plt.subplots(1, 2, figsize =  (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", ax = ax[0]);
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", ax = ax[1]);
```
