## Plotting multiple sets of data - Çoklu data ve legend verme

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2= [45372, 52345, 54148, 94256, 16544, 45600, 96542, 98148]
```

## Legend lokasyon verme fontsize-grafik çizgi sitili

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary, 'b:', label = "SALARY")
plt.plot(age, salary_2, 'r-', label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(0.5, 0., 0.5, 0.5)) # Legend'a ince ayar yapmak icin bbox_to_anchor(axis x,axis y,figure,figure)
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc = "lower right") # Legend'a ince ayar yapmak icin loc string veya int ile belirleyebiliriz
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", linewidth=4, label = "SALARY")           # linewidth = Çizgi kalınlığı
plt.plot(age, salary_2, "r", linestyle= ":", label = "SALARY_2")     # linestyle = Çizgi sitili
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc= 0)
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", linewidth=4, label = "SALARY")           # linewidth = Çizgi kalınlığı
plt.plot(age, salary_2, linestyle= ":", color="green", label = "SALARY_2")     # linestyle = Çizgi sitili
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc= 0) # "best"
plt.title("SALARY BY AGE");
```

## Function Method and Object Oriented Methods

### Function Method

```python
age = np.arange(25,44,2.5)
salary = np.arange(40000,76000,5000)
salary_2= [45372, 52345, 54148, 94256, 16544, 45600, 96542, 98148]
```

### Multiple Plots

```python
x = np.arange(-50, 50)

y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)


fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)
plt.tight_layout();
plt.show()
```

### Pie Chart

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance=0.75, autopct ="%.2f")  # labeldistance= labelleri iceri alir 
                                                                    
plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode= [0.1, 0, 0, 0]
plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance=0.50,autopct ="%.1f", explode=myexplode) # explote = myexplode verdigimiz degerlere gore bir dilimi ayirir
plt.show()
```

```python
plt.figure(figsize = (10,8))

color_list = ["gold","yellowgreen","lightcoral","lightskyblue","lightgreen","pink"]

explode_list = [0.3, 0.3, 0.3, 0.3]

y = np.array([25, 35, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels = mylabels, labeldistance = 0.75, autopct = "%.1f", startangle = 90, 
        shadow = True, pctdistance = 0.6, colors = color_list, explode = explode_list,)

plt.legend(loc = (0))
```
