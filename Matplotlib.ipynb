## Plotting multiple sets ofa data - oklu data ve legend verme

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2= [45372, 52345, 54148, 94256, 16544, 45600, 96542, 98148]
```

## Legend lokasyon verme fontsize-grafik izgi sitili

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary, 'b:', label = "SALARY")
plt.plot(age, salary_2, 'r-', label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(0.5, 0., 0.5, 0.5)) # Legend'a ince ayar yapmak icin
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc = "lower right") # Legend'a ince ayar yapmak icin
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", linewidth=4, label = "SALARY")           # linewidth = izgi kalnl
plt.plot(age, salary_2, "r", linestyle= ":", label = "SALARY_2")     # linestyle = izgi sitili
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc= 0)
plt.title("SALARY BY AGE");
```

```python
plt.figure(figsize = (10, 5))
plt.plot(age, salary, "b-", linewidth=4, label = "SALARY")           # linewidth = izgi kalnl
plt.plot(age, salary_2, linestyle= ":", color="green", label = "SALARY_2")     # linestyle = izgi sitili
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc= 0) # "best"
plt.title("SALARY BY AGE");
```
